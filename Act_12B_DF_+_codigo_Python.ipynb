```python
# Algoritmo para determinar si un número es positivo o negativo (Ejecución x2)

# 1. Inicio
def determinar_signo_doble():
    # Estructura para repetir el proceso 2 veces
    for i in range(2):
        print(f"\n--- Ejecución número {i + 1} ---")

        try:
            # 2. Leer NÚM
            NUM = int(input("Ingrese el número (NÚM): "))

            # 3. Si NÚM >= 0
            if NUM >= 0:
                # Entonces Hacer R = "POSITIVO"
                R = "POSITIVO"
            else:
                # Si no Hacer R = "NEGATIVO"
                R = "NEGATIVO"

            # 4. Escribir "el número es", R
            print(f"el número es {R}")

        except ValueError:
            print("Error: Por favor ingrese un número entero.")

# 5. Fin
if __name__ == "__main__":
    determinar_signo_doble()
print("Programa realizado por David Barrera 1388")
```
